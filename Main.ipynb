```python
import polars as pl
import numpy as np
```

```python
df = pl.scan_csv("data/Titanic-Dataset.csv").lazy()
```

```python
df.select(
    pl.pearson_corr(
        pl.col('Survived'),
        pl.col('Age').cast(pl.Int64)
    )
).collect()
```

```python
df.groupby(['Pclass','Sex']).agg([
    pl.col('Survived').mean().alias('Survived Rate'),
    pl.col('Age').mean().alias('Age Mean'),
    pl.col('Fare').max().alias('minimum fare')
]).sort(by='Pclass').collect()
```

```python
df = df.with_column(
    (pl.col('Sex') == 'male')
)
```

```python
df.groupby(['Age']).agg([
    pl.count().alias('n'),
    pl.col('Survived').mean().alias('Survived Rate'),
    pl.col('Fare').mean().alias('Age Mean'),
]).sort(by='Age').collect().head(20)
```
